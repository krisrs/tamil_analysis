```python
uyir_count = []
for u in range(len(uyir)) :
    count = tamil_txt[0].count(uyir[u],0,len(tamil_txt[0])-1)
    uyir_count.append([uyir[u],count])
print(uyir_count)
```

```python
umey_count = []
for m in range(0,18):
    for n in range(0,12):
        all_count=0
        this_count = tamil_txt[0].count(umey[m][n],0,len(tamil_txt[0])-1)
        if n == 0:
            x_count = 0
            for x in range(1,12):
                count = tamil_txt[0].count(umey[m][x],0,len(tamil_txt[0])-1)
                x_count = x_count + count
            y_count = tamil_txt[0].count(mey[m],0,len(tamil_txt[0])-1)
            this_count = this_count - x_count
            this_count = this_count - y_count
        umey_count.append([n+1,m+1,uyir[n],mey[m],umey[m][n],this_count])
```

```python
sans_count = []
for m in range(0,6):
    for n in range(0,13):
        this_count = tamil_txt[0].count(sans[m][n],0,len(tamil_txt[0])-1)
        if n == 1:
            x_count = 0
            for x in range(2,13):
                count = tamil_txt[0].count(sans[m][x],0,len(tamil_txt[0])-1)
                x_count = x_count + count
            y_count = tamil_txt[0].count(sans[m][0],0,len(tamil_txt[0])-1)
            this_count = this_count - x_count
            this_count = this_count - y_count
        sans_count.append([sans[m][n],this_count])
```
